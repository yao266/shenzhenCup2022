```python
yyqx['时间'] = pd.to_datetime(yyqx['时间'],errors='coerce', format='%Y-%m-%d %H:%M:%S')
```

```python
yyqx['内容'] =[str(i).replace('唐山','').replace('微博','').replace('视频','').replace('L','').replace('警方','').replace('烧烤店','') for i in yyqx['内容']]
```

```python
wordclound = ['']*len(segTime1)
```

```python
yyqx.head()
yyqx['groupIndex'] = '0'
```

```python
yyqx.rename(columns={"sentiment":"情绪"},inplace=True)
```

```python
for i in range(len(yyqx['时间'])):
    print('第{}条'.format(i+1))
    tag = yyqx['时间'][i]
    for j in range(len(segTime1)):
        if tag >= datetime.strptime(segTime1[-1],'%Y-%m-%d %H:%M:%S'):
            yyqx['groupIndex'][i] = segTime1[-1][5:-3]+'至'+'07-29 23:40'
            break
        if  (tag >= datetime.strptime(segTime1[j],"%Y-%m-%d %H:%M:%S")) & (tag <= datetime.strptime(segTime1[j+1],"%Y-%m-%d %H:%M:%S")):
            # print("1:   "+segTime[j] +'至'+ segTime[j+1])
            yyqx['groupIndex'][i] = segTime1[j][5:-3] +'至'+ segTime1[j+1][5:-3]
            # print("2:   "+segTime[j] +'至'+ segTime[j+1])
            wordclound[j] = wordclound[j]+ ''.join(yyqx['内容'][i].split(' '))
            break
```

```python
import plotly.express as px
```

```python
subset = yyqx[['groupIndex','情绪']]
bardata = subset.groupby(['groupIndex','情绪'])['情绪'].count()
s1 = pd.DataFrame(bardata)
s1['时间段'] = '0'
s1['情感'] = '0'
timelist = list(yyqx['groupIndex'].unique())
for i in range(len(s1['情绪'])):
    s1['时间段'][i] = timelist[int(i/2 )]
    if i % 2 ==0:
        s1['情感'][i] = '消极'
    else:
        s1['情感'][i] = '积极'
import math
print("系统中个体数目：",s1['情绪'][0]+s1['情绪'][1])
s1['情绪'] = [math.log2(num) for num in s1['情绪']]

fig = px.bar(s1, x='时间段',y='情绪',color='情感', template="simple_white")
fig.update_layout(
    font_size = 13,
    title= dict(
        text = "唐山打人事件各时间段情绪分布图",
        x = 0.45,
        font_size = 24
    ),
    xaxis = dict(
        title="话题传播时间段",
        titlefont_size = 24,
        tickfont=dict(
            color="black",
            size=15,
            family="airlblack"

        ),
        tickangle=-45
        
    ),
    yaxis=dict(
        title="网民情绪分布",
        titlefont_size = 24,
        tickfont=dict(
            color="black",
            size=15,
        )
    ),
    legend = dict(
        orientation = "h" ,# 开启水平显示
        y = 1.2,
        font_size = 24
    )
)
fig.show()
```
